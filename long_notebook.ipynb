```python
import numpy as np
import matplotlib.pylab as plt
```

```python
x = np.arange(7000)
y1 = np.cos(2*np.pi*x/1000)
y2 = np.cos(2*np.pi*x/500)
y3 = np.cos(2*np.pi*(x + 500)/500)
y4 = np.cos(2*np.pi*(x + 250)/500)
y5 = np.cos(2*np.pi*(x - 250)/500)
```

```python
fig, ax = plt.subplots(5, 1, figsize=(16,10))
ax[0].plot(x, y1)
ax[1].plot(x, y1 + y2)
ax[2].plot(x, y1 + y3)
ax[3].plot(x, y1 + y4)
ax[4].plot(x, y1 + y5);
```
